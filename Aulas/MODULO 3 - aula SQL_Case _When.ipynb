Objetivo: Aprofundar o conhecimento de SQL, a partir dde funções que permitem uma análise mais detalhada e otimizada dos dados.

Sintaxe

```
CASE
  WHEN condicao1 THEN resultado1
  WHEN condicao2 THEN resultado2
  ELSE resultado3
END;
```

Obs: pode ter até 10 linhas de WHEN

## Exercício:

No catálogo da Conquer Sales, em **quantas categorias diferentes** na produtos na **cor cinza**?

```
SELECT
CASE WHEN UPPER(product_name) LIKE '%GRAY%'
	THEN 'SIM'
    ELSE 'NAO'
 END VALIDADOR,
 COUNT (DISTINCT category) as QTDE_CATEGORIAS
 FROM product
 GROUP BY 
 CASE WHEN UPPER(product_name) LIKE '%GRAY%'
	THEN 'SIM'
    ELSE 'NAO'
 END

```

## EXERCÍCIO PT 2:

Se agrupar as categorias entre **'Furniture', 'Office Supplies', e 'Others'**, quantos produtos com a cor cinza na descrição terá no total?

```
SELECT
CASE
	WHEN UPPER(category) = 'OFFICE SUPPLIES' THEN category
    when UPPER(category) = 'FURNITURE' THEN category
    ELSE 'Others'
    end TIPO_CATEGORIA,
 SUM (CASE WHEN UPPER(product_name) LIKE '%GRAY%' THEN 1 ELSE 0 END) VALIDA_CINZA
 FROM product
 GROUP BY
 CASE
	WHEN UPPER(category) = 'OFFICE SUPPLIES' THEN category
    when UPPER(category) = 'FURNITURE' THEN category
    ELSE 'Others'
  end
```
